```python
# --- Support Functions ---
# This section contains support functions used by the main code.

def read_input(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()
```

```python
# --- Part ONE ---
input = read_input('input.txt')

# get first number for each line and use it as the first digit
# get last number for each line and use it as the last digit
numbers = []
for line in input:
    n = [int(char) for char in line if char.isdigit()]
    if n:
        numbers.append(10* n[0] + n[-1])

result = sum(numbers)
print("Part ONE:", result)
```

```python
# --- Part TWO ---
input = read_input('input.txt')

# now some numbers can be written as words
# get first number for each line and use it as the first digit
# get last number for each line and use it as the last digit
num_words = {'one':1, 'two':2, 'three':3, 'four':4,
             'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9}

numbers = []

line_nr = 0
for line in input:
    oline = line
    # substitute first word-number with the number
    wnum = sorted([(line.find(w), w) for w in num_words.keys() if line.find(w) != -1])
    if wnum:
        idx, w = wnum[0]
        line = line[:idx]+str(num_words[w])+line[idx+len(w):]

    # substitute last word-number with the number
    wnum = sorted([(line.rfind(w), w) for w in num_words.keys() if line.rfind(w) != -1])
    if wnum:
        #print(wnum)
        idx, w = wnum[-1]
        line = line[:idx]+str(num_words[w])+line[idx+len(w):]

    n = [int(char) for char in line if char.isdigit()]
    numbers.append(10* n[0] + n[-1])
    print (line_nr, oline, "->", line  , "->", numbers[-1])
    line_nr += 1
    #print(oline, "->", line, "->", numbers[-1])

result = sum(numbers)
# print(numbers)
print("Part TWO:", result)
```
